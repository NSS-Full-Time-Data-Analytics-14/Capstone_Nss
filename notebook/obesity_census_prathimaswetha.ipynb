## Data Cleaning

```python
obesity_df
```

```python
obesity_df['Age'] = obesity_df['Age'].astype(int).round(2)   ## Age is rounded to 2 decimal place converted from Float to Int 
```

```python
obesity_df['Height'] = obesity_df['Height'].round(2)  ## Height  is rounded to 2 decimal place 
```

```python
obesity_df['Weight'] = obesity_df['Weight'].round(2)  ## Weight is rounded to 2 decimal place 
```

```python
obesity_df['Number_of_vegetables'] = obesity_df['Number_of_vegetables'].astype(int).round(2) ##  Vegetables is rounded to 2 decimal place converted from Float to Int  
```

```python
obesity_df['Main_meals'] = obesity_df['Main_meals'].astype(int).round(2) ##  Main_meals is rounded to 2 decimal place converted from Float to Int  
```

```python
obesity_df['Water_intake'] = obesity_df['Water_intake'].astype(int).round(2) ##  Water_intake  is rounded to 2 decimal place converted from Float to Int  
```

```python
obesity_df['Physical_activity'] = obesity_df['Physical_activity'].astype(int).round(2) ## Physical_activity  is rounded to 2 decimal place converted from Float to Int  
```

```python
obesity_df['Physical_activity'] = obesity_df['Physical_activity'].astype(int).round(2) ## Physical_activity  is rounded to 2 decimal place converted from Float to Int  
```

```python
obesity_df['Screen_time'] = obesity_df['Screen_time'].astype(int).round(2) ## Screen_time  is rounded to 2 decimal place converted from Float to Int  
```

```python
obesity_df
```

 ## H in M, W in KG, Water_intake In L, Screen_time in H, Vegetables & main_meals on daily basis, Physical_activity on Weekly basis
